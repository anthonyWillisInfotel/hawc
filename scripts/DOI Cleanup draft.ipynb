```python
import html
```

```python
duplicate_doi_ids = {}
tarnished_doi_ids = []
for doi_id in doi_identifiers:
    new_doi = html.unescape(doi_id.unique_id)
    if (constants.DOI_EXACT.fullmatch(new_doi) and new_doi not in doi_id_saved): #changing &gt;/&lt; made a difference, save new doi
        doi_id.unique_id = new_doi
        doi_id.save()
        doi_id_saved.append(new_doi)
    elif (constants.DOI_EXACT.fullmatch(new_doi) and new_doi != doi_id.unique_id): #removing &gt;/&lt; worked but theres an existing duplicate
        print(f"duplicate Identifier: {doi_id}")
        duplicate_doi_ids[new_doi] = doi_id
    elif (not constants.DOI_EXACT.fullmatch(new_doi)): #doi needs to be extracted/validated further
        new_doi = constants.DOI_EXTRACT.search(new_doi)
        if new_doi:
            new_doi = new_doi.group(0)
        if (new_doi and new_doi[len(new_doi)-1] == '.'): #remove period at end of doi if it's there
            new_doi = new_doi[:len(new_doi)-1]
        if (new_doi is None or not constants.DOI_EXACT.fullmatch(new_doi)):
            tarnished_doi_ids.append(doi_id)
            print(f"id: {doi_id}\tdirty doi: {doi_id.unique_id}\tnew doi: {new_doi}")
        else:
            if new_doi in doi_id_saved:
                print(f"duplicate Identifier: {doi_id}")
                duplicate_doi_ids[new_doi] = doi_id
            else:
                doi_id.unique_id = new_doi
                doi_id.save() 
                doi_id_saved.append(doi_id.unique_id)
        
        
```

```python
for dupe, old_doi in duplicate_doi_ids.items():
    for reference in Reference.objects.filter(identifiers__id=old_doi.id):
        print(f'reference: \t{reference}')
        print(f'identifiers before: \t{reference.identifiers.all()}')
        other_ids = reference.identifiers.exclude(id=old_doi.id)
        reference.identifiers.set(other_ids)
        reference.identifiers.add(Identifiers.objects.filter(unique_id=dupe)[0])
        print(f'identifiers after: \t{reference.identifiers.all()}')
    old_doi.delete()
            
```

```python
refs = Reference.objects.exclude(identifiers__database=4)
```

```python
#this one will try to create new doi identifiers for all references without one. 
#warning: takes a while
for ref in refs:
    for ids in ref.identifiers.all():
        doi = constants.DOI_EXTRACT.search(ids.content)
        if (doi):
            doi = doi.group(0)
            if (doi[len(doi)-1] == ','):
                doi = doi[:len(doi)-1]
            if (doi[len(doi)-1] == '"'):
                doi = doi[:len(doi)-1]
            if (doi[len(doi)-1] == '.'):
                doi = doi[:len(doi)-1]
            try:
                existingID = Identifiers.objects.get(unique_id=doi)
                ref.identifiers.add(existingID)
                #print(f'existing ID!\t{existingID}')
            except (Identifiers.DoesNotExist):
                newID = Identifiers(unique_id=doi, database=4)
                newID.save()
                ref.identifiers.add(newID)
```
